**Task 1:**

```python
num = input("Enter a number: ")
num_int = int(num)
num_float = float(num)
num_str = str(num)
num_complex = complex(num)

print(f"Float: {num_float} -> {type(num_float)}")
print(f"String: {num_str} -> {type(num_str)}")
print(f"Complex: {num_complex} -> {type(num_complex)}")

half = num_int // 2 * 2
if half == num_int:
    print(f"{num_int} is divisible by 2")
else:
    print(f"{num_int} is not divisible by 2")
```

**Task 3:**

```python
n = input("Enter a 4 digit number: ")
swapped = n[3]+n[2]+n[1]+n[0]

encrypted = ""
for digit in swapped:
    new_digit = (int(digit) + 7) % 10
    encrypted += str(new_digit)
print("Encrypted number:", encrypted)
```

**Task 5:**

```python
name = input("Enter your name: ")
year = input("Enter your birth year: ")
first3 = name[:3]
styled = first3[0].upper() + first3[1:].lower()
last2 = year[-2:]

symbols = ['@','#','%','&','*']
ascii_val = ord(name[0])
symbol = symbols[ascii_val % len(symbols)]

password = styled + last2 + symbol
print("Generated password: ", password)
```
