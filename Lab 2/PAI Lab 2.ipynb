```python
def valid(password):
    if len(password) < 8: return False
    hasLetter = hasDigit = hasSpecial = False
    specials = "@#$%"
    for char in password:
        if char.isalpha():
            hasLetter = True
        elif char.isdigit():
            hasDigit = True
        elif char in specials:
            hasSpecial = True
    return hasLetter and hasDigit and hasSpecial

password = input("Enter a password: ")
if valid(password):
    print(f"'{password}' is a valid password.")
else:
    print(f"'{password}' is not a valid password.")
```

**Task 2:**

```python
def add_mat(mat1, mat2):
    rows = len(mat1)
    cols = len(mat1[0])

    result = []
    for i in range(rows):
        result += [[0] * cols]
    for i in range(rows):
        for j in range(cols):
            result[i][j] = mat1[i][j] + mat2[i][j]
    return result

def mul_mat(mat1, mat2):
    rows = len(mat1)
    cols = len(mat2[0])
    common = len(mat2)
    result = []
    for i in range(rows):
        result += [[0] * cols]
    for i in range(rows):
        for j in range(cols):
            sum = 0
            for k in range(common):
                sum = sum + mat1[i][k] * mat2[k][j]
            result[i][j] = sum
    return result


mat_a = [[1, 5],[3, 8]]
mat_b = [[4, 2],[7, 6]]

print(f"Matrix Addition Result: {add_mat(mat_a, mat_b)}")
print(f"Matrix Multiplication Result: {mul_mat(mat_a, mat_b)}")
```

**Task 4:**

```python
nums = (-10, -5, -2, 4, 7, 12, 16)
n = len(nums)
nums = sorted(nums)

left = 0
right = n - 1
best_pair = (nums[left], nums[right])
min_sum = abs(nums[left] + nums[right])

while left < right:
    curr_sum = nums[left] + nums[right]

    if abs(curr_sum) < min_sum:
        min_sum = abs(curr_sum)
        best_pair = (nums[left], nums[right])

    if curr_sum < 0:
        left += 1
    else:
        right -= 1

print("Pair closest to 0 sum: ", best_pair)
```

**Task 5:**

```python
students = {}

while True:
    print("\n--- Student Management System ---")
    print("1. Add student")
    print("2. Update marks")
    print("3. Delete student")
    print("4. Find topper")
    print("5. Show all students")
    print("6. Exit")
    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        name = input("Enter name: ")
        marks = int(input("Enter marks: "))
        students[name] = marks
        print(f"{name} added with {marks} marks")
    elif choice == "2":
        name = input("Enter name to update: ")
        if name in students:
            marks = int(input("Enter new marks: "))
            students[name] = marks
            print(f"{name}'s marks updated to {marks}")
        else:
            print("Student not found")
    elif choice == "3":
        name = input("Enter name to delete: ")
        if name in students:
            del students[name]
            print(f"{name} deleted.")
        else:
            print("Student not found")
    elif choice == "4":
        if students:
            topper = max(students, key=students.get)
            print(f"Topper: {topper} with {students[topper]} marks")
        else:
            print("No students in list bruh")
    elif choice == "5":
        if students:
            print("\nStudent List:")
            for name, marks in students.items():
                print(f"{name}: {marks}")
        else:
            print("No students in list")
    elif choice == "6":
        print("Exiting...")
        break
    else:
        print("Invalid choice")
```

```python
countries = {"India": 1410000000, "China": 1200000000, "USA": 332000000, "Indonesia": 276000000, "Pakistan": 240000000, "Brazil": 215000000, "Nigeria": 218000000}

print("Top 3 most populated countries:")
for i in range(3):
    max_country = None
    max_population = -1

    for country, population in countries.items():
        if population > max_population:
            max_population = population
            max_country = country

    if max_country:
        print(f"{i+1}. {max_country} - {max_population}")
        del countries[max_country]
```
