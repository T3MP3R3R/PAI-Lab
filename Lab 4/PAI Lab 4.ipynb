```python
class Vehicle:
    def __init__(self, seating_cap):
        self.seating_cap = seating_cap

    def fare(self):
        return self.seating_cap * 100

class Bus(Vehicle):
    def fare(self):
        total_fare = super().fare()
        total_fare += total_fare * 0.10
        return total_fare

bus = Bus(50)
print("Total Bus Fare:",bus.fare())
```

**Task 2:**

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side * self.side

r = Rectangle(5, 3)
t = Triangle(4, 6)
s = Square(4)
print("Rectangle Area:", r.area())
print("Triangle Area:", t.area())
print("Square Area:", s.area())
```

```python
class Student:
    def __init__(self, id, name):
        self.id = id
        self.name = name
    def display_student(self):
        print("Student ID:", self.id)
        print("Name:", self.name)

class Marks(Student):
    def __init__(self, id, name, algo, data_sci, calculus):
        super().__init__(id, name)
        self.marks_algo = algo
        self.marks_data_sci = data_sci
        self.marks_calculus = calculus
    def display_marks(self):
        print("Algo Marks:", self.marks_algo)
        print("Data Science Marks:", self.marks_data_sci)
        print("Calculus Marks:", self.marks_calculus)

class Result(Marks):
    def __init__(self, student_id, name, algo, data_sci, calculus):
        super().__init__(student_id, name, algo, data_sci, calculus)
    def display_result(self):
        total = self.marks_algo + self.marks_data_sci + self.marks_calculus
        avg = total/3
        print("Total Marks:", total)
        print("Average Marks:", avg)

r = Result(101, "Ayaan", 85, 90, 80)
r.display_student()
r.display_marks()
r.display_result()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def calculateBonus(self):
        return 0

class Manager(Employee):
    def calculateBonus(self):
        return self.salary * 0.20
    def hire(self):
        print(f"{self.name} is hiring a new employee")

class Developer(Employee):
    def calculateBonus(self):
        return self.salary * 0.10
    def writeCode(self):
        print(f"{self.name} is writing code")

class SeniorManager(Manager):
    def calculateBonus(self):
        return self.salary * 0.30

m = Manager("Ali", 80000)
d = Developer("Zara", 60000)
s = SeniorManager("Rehan", 100000)

print(m.name, "Bonus:", m.calculateBonus())
print(d.name, "Bonus:", d.calculateBonus())
print(s.name, "Bonus:", s.calculateBonus())

m.hire()
d.writeCode()
```
