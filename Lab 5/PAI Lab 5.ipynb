```python
try:
    filename = input("Enter file name: ")
    with open(filename, "r") as file:
        text = file.read()
        chars = len(text)
        words = len(text.split())
        print("Total Characters:", chars)
        print("Total Words:", words)
except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print("Something went wrong:", e)
```

**Task 2:**

```python
try:
    filename = input("Enter file name: ")
    old_word = input("Enter word/phrase to replace: ")
    new_word = input("Enter new word/phrase: ")

    with open(filename, "r") as file:
        content = file.read()

    content = content.replace(old_word, new_word)

    with open(filename, "w") as file:
        file.write(content)

    print("All occurrences replaced successfully")
except FileNotFoundError:
    print("Error: File not found.")
except Exception as e:
    print("Something went wrong:", e)
```

**Task 3:**

```python
try:
    list1 = input("Enter first list: ").split(",")
    list2 = input("Enter second list: ").split(",")

    if len(list1) != len(list2):
        raise ValueError("Both lists must have same number of elements")

    my_dict = {}
    for i in range(len(list1)):
        my_dict[list1[i].strip()] = list2[i].strip()

    with open("dict_file.txt", "w") as file:
        for k, v in my_dict.items():
            file.write(f"{k}:{v}\n")

    print("Dictionary saved to dict_file.txt successfully")
except ValueError as ve:
    print("Value Error:", ve)
except Exception as e:
    print("Error:", e)
```

**Task 5:**

```python
import json

try:
    data = {'Ali':23,'Saad':24,'Salman':15,'Shams':25,'Sadiq':46,'Hammad':23}
    with open("data.json", "w") as f:
        json.dump(data, f)
    with open("data.json", "r") as f:
        loaded = json.load(f)

    max_age = max(loaded.values())
    max_people = [name for name, age in loaded.items() if age == max_age]

    print(f"Oldest Age: {max_age}")
    print(f"People with max age: {', '.join(max_people)}")
    print("\nPeople with same age:")
    checked = set()
    for name, age in loaded.items():
        if list(loaded.values()).count(age) > 1 and age not in checked:
            dupes = [n for n, a in loaded.items() if a == age]
            print(f"Age {age}: {', '.join(dupes)}")
            checked.add(age)

except Exception as e:
    print("Error:", e)
```

**Task 6:**

```python
def writeQues():
    try:
        sentence = input("Enter a sentence: ")
        if sentence.endswith("?"):
            with open("questions.txt", "a") as f:
                f.write(sentence + "\n")
            print("Question saved to questions.txt successfully")
        else:
            print("Not a question, so nothing saved")
    except Exception as e:
        print("Error:", e)

writeQues()
```

**Task 7:**

```python
def fix_file():
    try:
        with open("original.txt", "r") as f:
            content = f.read()

        # Simple logic example where I replace z with s
        fixed = ""
        for ch in content:
            if ch == "z":
                fixed += "s"
            else:
                fixed += ch

        with open("replaced.txt", "w") as f:
            f.write(fixed)
        print("File corrected successfully")
    except FileNotFoundError:
        print("File not found")
    except Exception as e:
        print("Error:", e)

fix_file()
```
