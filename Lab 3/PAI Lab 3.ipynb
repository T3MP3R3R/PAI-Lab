**Task 2:**

```python
def check_last(string):
    vowels = 'aeiouAEIOU'
    if not string:
        return "No string given"
    last_char = string[-1]
    if last_char.isalpha():
        if last_char in vowels:
            return "Last char's a vowel"
        else:
            return "Last char's a consonant"
    else:
        return "Last char's not an alphabet"

result = check_last("Hello")
print(result)
result = check_last("World")
print(result)
```

**Task 3:**

```python
def employee(name, salary=10000):
    tax = 0.02 * salary
    salary_after_tax = salary - tax
    print(f"Employee Name: {name}")
    print(f"Salary after tax: {salary_after_tax}")

employee("Ali", 20000)
employee("Saima")
```

```python
def is_palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

ans = is_palindrome("Racecar")
print(ans)
ans = is_palindrome("Swims")
print(ans)
```
